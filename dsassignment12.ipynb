```python
#ans3
'''The ArithmeticError class in Python represents errors that occur during arithmetic operations.
   It serves as a base class for more specific arithmetic-related exceptions. Two notable exceptions derived
   from ArithmeticError are ZeroDivisionError and FloatingPointError.'''
#1.ZeroDivisionError
try:
    a=5/0
except ZeroDivisionError as e:
    print(e)
```

```python
#2.FloatingPointError
try:
    result = 5.0 / 0.0
except FloatingPointError as e:
    print(e)
```

```python
#ans4
'''The LookupError class in Python is a base class for exceptions that occur when a key or index used to
   access a mapping or sequence (like dictionaries, lists, or tuples) is not found. It is a subclass of the 
   Exception class and serves as a general category for errors related to looking up values.'''
#example1
'''KeyError: Raised when a dictionary key is not found.'''
try:
    d={"key":"abhi",1:[2,3,4]}
    d["key2"]
except KeyError as e:
    print(e)
    
```

```python
#example2
'''IndexError: Raised when a sequence subscript is out of range.'''
try:
    l=[1,2,3,4,5]
    print(l[6])
except IndexError as e:
    print(e)
```
